# Exploratory Data Analysis

```python
plot1 = sns.countplot(x = 'type_school', hue = 'in_college', data= go_to_college_data)
```

More students after academic school choose to continue to college in comparison to students after vocational school.

```python
plot2 = sns.countplot(x = 'school_accreditation', hue = 'in_college', data= go_to_college_data)
```

It is more likely that a student after A school will choose college.

```python
plot3 = sns.catplot(x = 'type_school', hue = 'in_college', col = 'school_accreditation', data= go_to_college_data, kind = 'count')
```

The same number of students in academic and vocational school with A accreditation go to college. In schools with B accreditation more students after academic type of school will go to college than after vocational type of school. 

```python
plot4 = sns.countplot(y = 'interest', hue = 'in_college', data= go_to_college_data)
```
