# Exploratory Data Analysis

```python
import seaborn as sns
```

```python
plot1 = sns.countplot(x = 'type_school', hue = 'in_college', data= go_to_college_data)
```

More students after academic school choose to continue to college in comparison to students after vocational school.

```python
plot2 = sns.catplot(x = 'type_school', hue = 'in_college', col = 'gender', data= go_to_college_data, kind = 'count')
```

More female than male students after academic type of school decide to go to college.
